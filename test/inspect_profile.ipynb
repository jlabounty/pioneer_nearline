```python
import hist
import pandas 
import matplotlib.pyplot as plt 
import matplotlib
import numpy as np
```

```python
infile = '../data/run45/data/subrun0/profile.csv'
df = pandas.read_csv(infile, header=None)
df = df.loc[df[5] < 1e9]
df.head()
```

```python
for x in df.columns:
    plt.plot(df[x])
    plt.show()
```

```python
def get_mean_std(profile):
    mean_x = np.average(df[0], weights=df[2])
    std_x = np.sqrt( np.average((df[0]-mean_x)**2, weights=df[2]) )
    mean_y = np.average(df[1], weights=df[2])
    std_y = np.sqrt( np.average((df[1]-mean_y)**2, weights=df[2]) )

    return (
        (
            mean_x,
            std_x
        ),
        (
            mean_y,
            std_y
        ),
    )
```

```python
plt.scatter(df[0], df[1], c=df[2])
ding = get_mean_std(df)
plt.title(str(ding))
# plt.ylim(-100,100)
plt.colorbar()
```
